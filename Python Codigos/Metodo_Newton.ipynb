```python
import sympy as sp
import math
import numpy as np

def gradient(f, var):
    n = len(var)
    grad = sp.zeros(n, 1)
    for i in range(n):
        grad[i] = f.diff(var[i])
    return grad
    
""" Nuestra problema a resolver"""
x, y = sp.symbols("x y")
f = -sp.sin(x * y) + sp.cos(x - y)  # Función a maximizar
f = -f  # Convertir en función a minimizar

print(f"Minimizar f = {f}")

""" Hallamos el gradiente y la hessiana de f"""
grad = gradient(f, (x, y))
H = sp.hessian(f, (x, y))
H_inv = H.inv()

""" Iteramos """
pi_value = math.pi
x0 = sp.Matrix([pi_value, 0])
tolerancia = 1e-10
max_iteraciones = 100

for i in range(max_iteraciones):
    print(f"\nITERACIÓN {i+1}")
    print(f"Punto x{i} = {x0.evalf()}")

    # Evaluamos en el punto x0 las matrices
    H_inv_eval = H_inv.subs([(x, x0[0]), (y, x0[1])])
    grad_eval = grad.subs([(x, x0[0]), (y, x0[1])])

    # Obtenemos el siguiente punto x1
    x1 = x0 - H_inv_eval * grad_eval
    print(f"Punto x{i+1} = {x1.evalf()}")
    
    # Evaluamos la normal para la tolerancia
    x_prev = np.array([float(x0[0]), float(x0[1])])
    x_new = np.array([float(x1[0]), float(x1[1])])

    if np.linalg.norm(x_prev - x_new) < tolerancia:
        print(f"\nConvergencia alcanzada en la iteración {i+1}")
        break

    x0 = x1

print(f"\nSolución encontrada usando el método de Newton: {x0.evalf()}")
```
