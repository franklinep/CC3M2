```python
import numpy as np
from sympy import *

# variables x1 y x2
x, y = symbols('x y')

# Nuestra funcion a maximizar
f = 4*x + 6*y - 2*x**2 - 2*x*y - 2*y**2
print(f"Funcion objetiva: ", f)
# Cambiamos a un problema de minimizacion
f = -f

# Obtenemos sus derivadas parciales dfx1, dfx2
dfx1 = diff(f, x)
dfx2 = diff(f, y)

# Nuestra funcion gradiente descendiente
def gradient_descent(x0, y0, tol=1e-4, max_iters=1000):
    
    i = 0
    while i < max_iters:
        print(f"\nITERACION {i+1}")
        print(f"Punto x_{i} = ({x0}, {y0})")
        # theta (t)
        t = symbols('t')
        # Hallamos el minimo g(t), theta minimo
        g = f.subs({x : x0 + t * dfx1.subs({x : x0, y : y0}),
                    y : y0 + t * dfx2.subs({x : x0, y : y0})})
        # Metodo de la primera derivada para hallar el minimo g
        dgt = diff(g,t)
        print("Funcion g(t) = ", g)
        puntos_criticos = solve(dgt, t) 
        print(f"Puntos criticos = {puntos_criticos}")
        # Evalúa la función en los puntos críticos
        t_min = 0
        g_min = oo
        for p in puntos_criticos:
            if (g.subs(t, p) < g_min):
                t_min = p
                g_min = g.subs(t,p)
        print(f"theta minimo = ", t_min)
        # Verifica que sea la Segunda derivada positiva
        if(diff(dgt, t) <= 0): break;
        
        # Obtenemos el siguiente punto
            
        x1 = x0 + t_min * dfx1.subs({x : x0, y : y0})
        y1 = y0 + t_min * dfx2.subs({x : x0, y : y0})

        # Evaluamos la normal para la tolerancia
        x_prev = np.array([float(x0),float(y0)])
        x_new = np.array([float(x1),float(y1)])
        
        if (np.linalg.norm(x_prev - x_new) < tol):
            break

        x0, y0 = x1, y1
        i += 1

    return [x1,y1]

x0 = y0 = 1
x = gradient_descent(x0,y0)
print(f"La solución aproximada es:\n({x[0], x[1]}) = ({float(x[0])}, {float(x[1])})")
```
