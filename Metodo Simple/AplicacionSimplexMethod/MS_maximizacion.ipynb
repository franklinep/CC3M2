
## METODO SIMPLEX (MAXIMIZACION)

> **EJEMPLO 01**



\begin{align*}
\text{Maximizar:} \quad &3x_1 + 5x_2 \\
\text{sujeto a:} \quad &214x_1 + 616x_2 &\leq 4 \\
&589x_1 + 209x_2 &\leq 5.5 \\
&55x_1 + 33x_2 &\leq 7 \\
&x_1, x_2 &\geq 0
\end{align*}


\begin{align*}
\quad &z - 3x_1 - 5x_2 = 0\\
\text{Tenemos que:} \quad &214x_1 + 616x_2 + s_1 &= 4 \\
&589x_1 + 209x_2 + s_2 &= 5.5 \\
&55x_1 + 33x_2 + s_3 &= 7 \\
&x_1, x_2 &\geq 0
\end{align*}


```python
import pandas as pd
import numpy as np
import copy

from fractions import *
from tabulate import *

np.set_printoptions(precision=9)  # Establecer precisi√≥n a 9 decimales


def print_table(data, n, m):
    hdrs = [' ','z'] + [f'x{i}' for i in range(0, n)] + [f's{i}' for i in range(0, m)] + ['r']
    print(tabulate(data, hdrs, 'pretty'))
    print()

def print_table2(data, v, n, m):
    hdrs = [' ','z'] + [f'x{i}' for i in range(0, n)] + [f's{i}' for i in range(0, m)] + ['r', 'CR/piv']
    data2 = copy.deepcopy(data)
    for i in range (0,m):
        data2[i].append(v[i])
    print(tabulate(data2, hdrs, 'pretty'))
    print()

def checkStop(v):
    for e in v:
        if e < 0:
            return False
    return True
def get_CR_piv(data, idx_col_piv):
    CR_piv = []
    for i in range(0,m):
        num = data[i][-1]
        den = data[i][idx_col_piv]
        if den == 0:
            CR_piv.append(np.inf * np.sign(num))
        else:
            CR_piv.append(data[i][-1] / data[i][idx_col_piv])
    return CR_piv

def OEF(data, i, c, k): # fi - (c)fk
    #print(f'=> Hacemos {data[0][i]} - ({c})x{idx_col_piv-2}')
    for j in range(1,len(data[0])):
        data[i][j] = data[i][j] - c * data[k][j]

#====================================================================================================
#====================================================================================================
#====================================================================================================

# n = numero de var NB | m = numero de var B
n = 2
m = 3
# TABLA DE DATOS
data = [
    ['s0', 0, 214, 616, 1, 0, 0, 4],
    ['s1', 0, 589, 209, 0, 1, 0, 5.5],
    ['s2', 0, 55, 33, 0, 0, 1, 7],
    ['z', 1, -3, -5, 0, 0, 0, 0]
]
print('TABLA INICIAL')
print_table(data,n,m)

k = 0


while not checkStop(data[m][1:]):

    print('\n\n\tITERACION : k =',k)
    # Indice de la columna pivote
    idx_col_piv = data[m][1:].index(min(data[m][1:])) + 1

    # Columa de coef tras dividir la col piv
    CR_piv = get_CR_piv(data, idx_col_piv)

    # Indice de la fila pivote (menor positivo)
    idx_fil_piv = CR_piv.index(min(CR_piv, key = lambda x: np.inf if x < 0 else x))

    # Elemento pivote
    piv = data[idx_fil_piv][idx_col_piv]



    print_table2(data,CR_piv, n, m)
    print('=======================')
    print('Columna pivote : ' + f'x{idx_col_piv-2}' )
    print('Fila pivote : ' + data[idx_fil_piv][0])
    print('Elemento Pivote :', piv)
    print('=======================\n')

    # Cambio de si a xi
    print('=> Se cambia la variable de ' + data[idx_fil_piv][0] + ' a ' + f'x{idx_col_piv-2}' + ' y hacemos al pivote 1')
    data[idx_fil_piv][0] = f'x{idx_col_piv-2}' # VB|z

    # Fila del pivote por (1/piv)
    for i in range(1,n+m+3):
        data[idx_fil_piv][i] = data[idx_fil_piv][i] / piv

    print_table(data,n,m)

    # Hacer 0's los elementos de la misma col del pivote

    print('=> Haciendo OEF hacemos los demas elem de la col del piv igual a 0')

    for i in range(0,m+1):
        if (i == idx_fil_piv):
            continue
        value = data[i][idx_col_piv]
        OEF(data,i,value,idx_fil_piv)

    print_table(data,n,m)
    k += 1

print('Por lo tanto los valores optimos son: ')
for i in range(0,m):
    if data[i][0][0] == 's':
        continue
    print(f'{data[i][0]} = {data[i][-1]}' )

x1 = data[0][-1]
x0 = data[1][-1]
#x00 = 0.006432907
#x11 = 0.009332341

print(214 * x0 + 616 * x1)
print(589 * x0 + 209 * x1)
print(55 * x0 + 33 * x1)


```

\begin{align*}
\text{Maximizar:} \quad &3x_1 + x_2 + 3x_3 \\
\text{sujeto a:} \quad &2x_1 + x_2 + x_3 &\leq 2 \\
&x_1 + 2x_2 + 3x_3 &\leq 5 \\
&2x_1 + 2x_2 + x_3 &\leq 6 \\
&x_1, x_2, x_3 &\geq 0
\end{align*}


```python
import pandas as pd
import numpy as np
import copy

from fractions import *
from tabulate import *

def print_table(data, n, m):
    hdrs = [' ','z'] + [f'x{i}' for i in range(0, n)] + [f's{i}' for i in range(0, m)] + ['r']
    print(tabulate(data, hdrs, 'pretty'))
    print()

def print_table2(data, v, n, m):
    hdrs = [' ','z'] + [f'x{i}' for i in range(0, n)] + [f's{i}' for i in range(0, m)] + ['r', 'CR/piv']
    data2 = copy.deepcopy(data)
    for i in range (0,m):
        data2[i].append(v[i])
    print(tabulate(data2, hdrs, 'pretty'))
    print()

def df(a, b):
    return(Fraction(a,b))

def checkStop(v):
    for e in v:
        if e < 0:
            return False
    return True
def get_CR_piv(data, idx_col_piv):
    CR_piv = []
    for i in range(0,m):
        num = data[i][-1]
        den = data[i][idx_col_piv]
        if den == 0:
            CR_piv.append(np.inf * np.sign(num))
        else:
            CR_piv.append(data[i][-1] / data[i][idx_col_piv])
    return CR_piv

def OEF(data, i, c, k): # fi - (c)fk
    #print(f'=> Hacemos {data[0][i]} - ({c})x{idx_col_piv-2}')
    for j in range(1,len(data[0])):
        data[i][j] = data[i][j] - c * data[k][j]



#####################################
# MAX: 3x1 + x2 + 3x3
# 2x1 + x2 + x3 <= 2
# x1 + 2x2 + 3x3 <= 5
# 2x1 + 2x2 + x3 <= 6

# x1,x2,x3 >= 0
#####################################

# n = numero de var NB | m = numero de var B
n = m = 3
# TABLA DE DATOS
data = [
    ['s0', 0, 2, 1, 1, 1, 0, 0, 2],
    ['s1', 0, 1, 2, 3, 0, 1, 0, 5],
    ['s2', 0, 2, 2, 1, 0, 0, 1, 6],
    ['z', 1, -3, -1, -3, 0, 0, 0, 0],
]
print('TABLA INICIAL')
print_table(data,n,m)

k = 0

while not checkStop(data[n][1:]) and k < 100:

    print('\n\n\tITERACION : k =',k)
    # Indice de la columna pivote
    idx_col_piv = data[m][1:].index(min(data[m][1:])) + 1

    # Columa de coef tras dividir la col piv
    CR_piv = get_CR_piv(data, idx_col_piv)

    # Indice de la fila pivote
    idx_fil_piv = CR_piv.index(min(CR_piv, key = lambda x: np.inf if x < 0 else x))

    # Elemento pivote
    piv = data[idx_fil_piv][idx_col_piv]



    print_table2(data,CR_piv, n, m)
    print('=======================')
    print('Columna pivote : ' + f'x{idx_col_piv-2}' )
    print('Fila pivote : ' + data[idx_fil_piv][0])
    print('Elemento Pivote :', piv)
    print('=======================\n')

    # Cambio de si a xi
    print('=> Se cambia la variable de ' + data[idx_fil_piv][0] + ' a ' + f'x{idx_col_piv-2}' + ' y hacemos al pivote 1')
    data[idx_fil_piv][0] = f'x{idx_col_piv-2}' # VB|z

    # Fila del pivote por (1/piv)
    for i in range(1,n+m+3):
        data[idx_fil_piv][i] = df(data[idx_fil_piv][i], piv)

    print_table(data,n,m)

    # Hacer 0's los elementos de la misma col del pivote

    print('=> Haciendo OEF hacemos los demas elem de la col del piv igual a 0')

    for i in range(0,m+1):
        if (i == idx_fil_piv):
            continue
        value = data[i][idx_col_piv]
        OEF(data,i,value,idx_fil_piv)

    print_table(data,n,m)
    k += 1

print('Por lo tanto los valores optimos son: ')
for i in range(0,m):
    if data[i][0][0] == 's':
        continue
    print(f'{data[i][0]} = {data[i][-1]}' )
```
